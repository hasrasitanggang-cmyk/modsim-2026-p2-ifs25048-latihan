```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setup style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 11

# Load data
df = pd.read_excel('data_kuesioner.xlsx', sheet_name='Kuesioner')
question_cols = [col for col in df.columns if str(col).startswith('Q')]
df_questions = df[question_cols].astype(str)

# Flatten dan bersihkan data
all_responses = df_questions.values.flatten()
all_responses = all_responses[~pd.isna(all_responses) & (all_responses != 'nan')]

# Hitung distribusi
response_order = ['SS', 'S', 'CS', 'CTS', 'TS', 'STS']
response_counts = pd.Series(all_responses).value_counts().reindex(response_order, fill_value=0)

# Visualisasi
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=response_counts.index, y=response_counts.values, palette="viridis", edgecolor='black')

plt.title('Distribusi Jawaban Keseluruhan (113 Partisipan Ã— 17 Pertanyaan)', 
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Jenis Respons', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)

# Tambahkan label nilai
for i, v in enumerate(response_counts.values):
    ax.text(i, v + 25, str(int(v)), ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
```
